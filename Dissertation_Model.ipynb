```python
# Function to translate the label into defect types 
def read_label(label, defect_types =''):
    if np.sum(label) == 0:
        defect_types = 'Normal wafer'
    else:
        for digit in range(np.shape(label)[0]):
            # Digit will also range from 0 to 8
            if label[digit] == 1:
                defect_types = defect_types + label_keys [digit]

    return defect_types
```

```python
# Removes duplicate 2D NumPy arrays from a list of wafer maps.
def remove_duplicate_wafer_maps(wafer_maps):
    seen_hashes = set()
    unique_maps = []
    
    for wafer in wafer_maps:
        # Convert the array to a bytes representation for hashing
        wafer_bytes = wafer.tobytes()
        wafer_hash= hash (wafer_bytes)
        
        if wafer_hash not in seen_hashes:
            seen_hashes.add(wafer_hash)
            unique_maps.append(wafer)
            
    return unique_maps

unique_maps = remove_duplicate_wafer_maps (train)
print(len(unique_maps))
```

```python
def build_cnn_model(input_shape, num_classes, trainable=True):
    inputs = Input(shape=input_shape)

    # Conv Block 1
    x = Conv2D(32, (3, 3), strides=(2, 2), padding='same', trainable=trainable)(inputs)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Conv Block 2
    x = Conv2D(64, (3, 3), strides=(2, 2), padding='same', trainable=trainable)(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Conv Block 3
    x = Conv2D(128, (3, 3), strides=(2, 2), padding='same', trainable=trainable)(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Conv Block 4
    x = Conv2D(256, (3, 3), padding='same', trainable=trainable)(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Conv Block 5
    x = Conv2D(128, (3, 3), strides=(2, 2), padding='same', trainable=trainable)(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    if num_classes is None:
        # Return feature maps if no classification head is requested
        return inputs, x
    else:
        # Global Pooling
        x = GlobalAveragePooling2D()(x)

        # Fully Connected Layer
        x = Dense(128, activation='relu', trainable=trainable)(x)
        x = Dropout(0.5)(x)

        # Output Layer
        outputs = Dense(num_classes, activation='softmax', trainable=trainable)(x)

        return inputs, outputs
```

```python
# custom callback class to track and visualize training progress
class LossHistory(keras.callbacks.Callback):
    # Called at beginning of training
    def on_train_begin(self, logs=None):
        self.losses = {'batch': [], 'epoch': []}
        self.accuracy = {'batch': [], 'epoch': []}
        self.val_loss = {'batch': [], 'epoch': []}
        self.val_accuracy = {'batch': [], 'epoch': []}

    # Called at end of each batch
    def on_batch_end(self, batch, logs=None):
        self.losses['batch'].append(logs.get('loss'))
        self.accuracy['batch'].append(logs.get('accuracy'))
        self.val_loss['batch'].append(logs.get('val_loss'))
        self.val_accuracy['batch'].append(logs.get('val_accuracy'))

    # Called at end of each epoch
    def on_epoch_end(self, epoch, logs=None):
        self.losses['epoch'].append(logs.get('loss'))
        self.accuracy['epoch'].append(logs.get('accuracy'))
        self.val_loss['epoch'].append(logs.get('val_loss'))
        self.val_accuracy['epoch'].append(logs.get('val_accuracy'))

    # Helper method to draw and save plots for given metric
    def draw_plot(self, data_list, label, scope):
        plt.figure()
        plt.plot(range(len(data_list)), data_list, label=label)
        plt.ylabel(label)
        plt.xlabel(scope)
        plt.legend(loc="upper right")
        plt.savefig(fr'C:\Users\Amrita Chaudri\Downloads\Sem 4\CNN Model File\Images\image_{label}_{scope}.jpg')
        plt.close()

    # Method to generate and save all plots at the end of training 
    def end_draw(self):
        self.draw_plot(self.losses["batch"], 'Loss', 'Train_Batch')
        self.draw_plot(self.accuracy["batch"], 'Accuracy', 'Train_Batch')
        self.draw_plot(self.val_loss["batch"], 'Val_Loss', 'Val_Batch')
        self.draw_plot(self.val_accuracy["batch"], 'Val_Accuracy', 'Val_Batch')
        self.draw_plot(self.losses["epoch"], 'Loss', 'Train_Epoch')
        self.draw_plot(self.accuracy["epoch"], 'Accuracy', 'Train_Epoch')
        self.draw_plot(self.val_loss["epoch"], 'Val_Loss', 'Val_Epoch')
        self.draw_plot(self.val_accuracy["epoch"], 'Val_Accuracy', 'Val_Epoch')
```

```python
(_1,_,image_size, b, c) = np.shape(X_train) 
input_shape = (image_size, image_size, c)
label_size = np.shape(y_test)[1]
patch_size = 13
num_patches = (image_size // patch_size) ** 2
projection_dim = 96
num_heads = 4
transformer_units = [projection_dim * 2, projection_dim,]  # Size of the transformer layers
transformer_layers = 16
mlp_head_units = [2048, 1024]
```

```python
def augmentation():
        # This funtion is as object function to do data augmentation
        data_augmentation = keras.Sequential(
            [
                tf.keras.layers.experimental.preprocessing.Normalization(),
                tf.keras.layers.experimental.preprocessing.RandomRotation(factor=0.02),
                tf.keras.layers.experimental.preprocessing.RandomZoom(height_factor=0.2, width_factor=0.2),
            ], name="data_augmentation",)

        # Compute the mean and the variance of the training data for normalization.
        data_augmentation.layers[0].adapt(X_train)
```

```python
def mlp(x, hidden_units, dropout_rate):
    # Multilayer Perceptron (MLP) head
    for units in hidden_units:
        x = Dense(units, activation=tf.nn.gelu)(x)
        x = Dropout(dropout_rate)(x)
    return x
```

```python
# Function to create a custom keras layer that extracts non overlapping patches from input image
def get_patches_layers(patch_size):
    return layers.Lambda(lambda images: tf.reshape(
        tf.image.extract_patches(
            images=images,
            sizes=[1, patch_size, patch_size, 1],
            strides=[1, patch_size, patch_size, 1],
            rates=[1, 1, 1, 1],
            padding = "VALID"
        ),
        [tf.shape(images)[0], -1, (patch_size*patch_size*images.shape[-1])]
    ))
```

```python
# function to encode image patches using a linear projection and positional embedding
def get_patchencoder(patch):

    projection = Dense(units=projection_dim)
    position_embedding = layers.Embedding(input_dim=num_patches, output_dim=projection_dim)

    positions = tf.range(start=0, limit=num_patches, delta=1)
    encoded = projection(patch) + position_embedding(positions)

    return encoded
```

```python
def create_vit_model(input_shape):
        inputs = Input(input_shape)

        # Augment data.
        images = inputs #data_augmentation(inputs)

        # Create patches.
        patches = get_patches_layers(patch_size)(images)

        # Encode patches.
        encoded_patches = get_patchencoder(patches)

        #print(np.shape(encoded_patches))

        # Create multiple layers of the Transformer block.

        for _ in range(transformer_layers):

            # Layer normalization 1.
            x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)

            # Create a multi-head attention layer.
            attention_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim, dropout=0.1)(x1, x1)

            #print(np.shape(attention_output))

            # Skip connection 1.
            x2 = layers.Add()([attention_output, encoded_patches])

            # Layer normalization 2.
            x3 = layers.LayerNormalization(epsilon=1e-6)(x2)

            # MLP.
            x3 = mlp(x3, hidden_units=transformer_units, dropout_rate=0.1)

            # Skip connection 2.
            encoded_patches = layers.Add()([x3, x2])

        #print('encoded_patches:', np.shape(encoded_patches))

        # Create a [batch_size, projection_dim] tensor.
        representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        representation = Flatten()(representation)
        representation = Dropout(0.2)(representation)

        #print('representation:', np.shape(representation))

        # Add MLP.
        features = mlp(representation, hidden_units=mlp_head_units, dropout_rate=0.5)

        #print('features:', np.shape(features))

        # Classify outputs.
        #features = layers.Dense(label_size*2, kernel_initializer='he_uniform', activation='relu')(features)
        logits = Dense(label_size, activation='sigmoid')(features)

        # Create the Keras model.
        model = Model(inputs=inputs, outputs=logits)

        print(model.summary())

        return model
```

```python
def train_model(batch_size = 32, num_epochs = 50, save_path='none'):
        
        model_vit = create_vit_model(input_shape)
        
        if save_path != 'none':
            print("set model-saving: " + save_path )
            print("set history-saving: " + save_path+'_history' )
        
        lr_schedule = keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=1e-3, decay_steps=10000, decay_rate=0.9)
        opt = tf.keras.optimizers.Adam(lr_schedule)

        model_vit.compile(optimizer=opt, loss='binary_crossentropy',metrics=[keras.metrics.BinaryAccuracy(name="accuracy")])

        checkpoint_filepath = "./tmp/checkpoint.weights.h5"
        checkpoint_callback = keras.callbacks.ModelCheckpoint(
            checkpoint_filepath,
            monitor="val_accuracy",
            save_best_only=True,
            save_weights_only=True,
        )

        history = model_vit.fit(
            x=X_train,
            y=y_train,
            batch_size=batch_size,
            epochs=num_epochs,
            validation_split=0.1,
            callbacks=[checkpoint_callback],
        )

        model_vit.load_weights(checkpoint_filepath)
        loss, accuracy = model_vit.evaluate(X_test, y_test)
        print(f"Test accuracy: {round(accuracy * 100, 2)}%")
        
        if save_path!= 'none':
            model_vit.save_weights(save_path)
            
            with open(save_path+'_history', 'wb') as file_pi:
                pickle.dump(history.history, file_pi)
        
        return history
```

```python
def load_model(path, plot=False):            
        try:
            del model_vit
        except:
            print("load model")
        
        model_vit = create_vit_model(input_shape)
        model_vit.load_weights(path)
        lr_schedule = keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=1e-3, decay_steps=10000, decay_rate=0.9)
        opt = tf.keras.optimizers.Adam(lr_schedule)

        model_vit.compile(optimizer=opt, loss='binary_crossentropy',metrics=keras.metrics.BinaryAccuracy(name="accuracy"))
        
        with open(path+'_history', "rb") as file_pi:
            history = pickle.load(file_pi)
        
        if (plot):
            # Plot training and validation loss
            # Plot training and validation accuracy
            plt.figure(figsize=(12, 4))

            plt.subplot(1, 2, 1)
            plt.plot(history['accuracy'], label='Training Accuracy')
            plt.plot(history['val_accuracy'], label='Validation Accuracy')
            plt.title('Model Accuracy')
            plt.xlabel('Epoch')
            plt.ylabel('Accuracy')
            plt.legend()

            # Plot training and validation loss
            plt.subplot(1, 2, 2)
            plt.plot(history['loss'], label='Training Loss')
            plt.plot(history['val_loss'], label='Validation Loss')
            plt.title('Model Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            plt.show()
        return model_vit
```

```python
def model_eval(model, x_test, y_test):
        # Evaluate the model
        loss, accuracy = model.evaluate(x_test, y_test)
        print(f"Loss: {loss:.4f}")
        print(f"Accuracy: {accuracy:.4f}")
```

```python
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, Reshape, LayerNormalization, Dense, Dropout, GlobalAveragePooling1D,Add,MultiHeadAttention
from tensorflow.keras.models import Model
```

```python
def build_hybrid_cnn_vit(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # CNN Feature Extractor
    x = Conv2D(32, (3, 3), strides=(2, 2), padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(64, (3, 3), strides=(2, 2), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, (3, 3), strides=(2, 2), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, (3, 3), strides=(2, 2), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, (3, 3), strides=(2, 2), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Reshape CNN outputs to patches
    patch_dim = x.shape[1] * x.shape[2]  # Get the last dimension (channels)
    patches = Reshape((patch_dim,x.shape[-1]))(x) # Reshape to (num_patches, patch_dim)

    # Transformer Encoder
    for _ in range(4):
        x1 = LayerNormalization(epsilon=1e-6)(x)
        attention_output = MultiHeadAttention(num_heads=4, key_dim=64)(x1, x1) # Assuming MultiHeadAttention is defined elsewhere
        x2 = Add()([attention_output, x1])
        x3 = LayerNormalization(epsilon=1e-6)(x2)
        x3 = Dense(128, activation='gelu')(x3)
        x3 = Dropout(0.1)(x3)
        x = Add()([x2, x3])

    # Classification Head
    x = Flatten()(x)
    x = Dropout(0.3)(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(num_classes, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model
```

```python
def train_hybrid_model(batch_size=32, num_epochs=50, save_path='none'):
    model = build_hybrid_cnn_vit(input_shape=X_train.shape[1:], num_classes=y_train.shape[-1])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss='binary_crossentropy', metrics=['accuracy'])

    checkpoint_filepath = './tmp/hybrid_checkpoint_weights.h5'
    checkpoint_callback = keras.callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath,
        save_weights_only=True,
        monitor='val_accuracy',
        save_best_only=True
    )

    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        batch_size=batch_size,
                        epochs=num_epochs,
                        callbacks=[checkpoint_callback]
                        )
    model.load_weights(checkpoint_filepath) # Load best weights
    
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"Hybrid Model Test Accuracy: {accuracy:.4f}")

    if save_path != 'none':
        model.save_weights(save_path)
        with open(save_path + "_history.db", 'wb') as file_pi:
            pickle.dump(history.history, file_pi)

    return model, history
```

### Kernel Regularization

```python
from tensorflow.keras.regularizers import l2

def build_hybrid_cnn_vit_kernal(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # CNN Feature Extractor with L2 Regularization
    x = Conv2D(32, (3, 3), strides=(2, 2), padding='same',
               kernel_regularizer=l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(64, (3, 3), strides=(2, 2), padding='same',
               kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, (3, 3), strides=(2, 2), padding='same',
               kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, (3, 3), strides=(2, 2), padding='same',
               kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, (3, 3), strides=(2, 2), padding='same',
               kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Reshape CNN outputs to patches
    patch_dim = x.shape[1] * x.shape[2]
    patches = Reshape((patch_dim, x.shape[-1]))(x)

    # Transformer Encoder
    for _ in range(4):
        x1 = LayerNormalization(epsilon=1e-6)(patches)
        attention_output = MultiHeadAttention(num_heads=4, key_dim=64)(x1, x1)
        x2 = Add()([attention_output, x1])
        x3 = LayerNormalization(epsilon=1e-6)(x2)
        x3 = Dense(128, activation='gelu', kernel_regularizer=l2(1e-4))(x3)
        x3 = Dropout(0.1)(x3)
        patches = Add()([x2, x3])

    # Classification Head with Regularization
    x = Flatten()(patches)
    x = Dropout(0.3)(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.3)(x)
    outputs = Dense(num_classes, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

def train_hybrid_model(batch_size=32, num_epochs=50, save_path='none'):
    model = build_hybrid_cnn_vit_kernal(input_shape=X_train.shape[1:], num_classes=y_train.shape[-1])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    checkpoint_filepath = './tmp/hybrid_checkpoint_weights.h5'
    checkpoint_callback = keras.callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath,
        save_weights_only=True,
        monitor='val_accuracy',
        save_best_only=True
    )

    # Dynamic Learning Rate Scheduler
    lr_scheduler = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )

    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        batch_size=batch_size,
                        epochs=num_epochs,
                        callbacks=[checkpoint_callback, lr_scheduler]
                        )

    model.load_weights(checkpoint_filepath)  # Load best weights

    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"Hybrid Model Test Accuracy: {accuracy:.4f}")

    if save_path != 'none':
        model.save_weights(save_path)
        with open(save_path + "_history.db", 'wb') as file_pi:
            pickle.dump(history.history, file_pi)

    return model, history
```

## Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
def train_hybrid_model(batch_size=32, num_epochs=50, save_path='none'):
    # Build the hybrid CNN-ViT model
    model = build_hybrid_cnn_vit(input_shape=X_train.shape[1:], num_classes=y_train.shape[-1])
    
    # Optimizer: Adam with initial learning rate
    optimizer = Adam(learning_rate=1e-3)
    
    model.compile(optimizer=optimizer,
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Checkpoint to save best weights
    checkpoint_filepath = './tmp/hybrid_checkpoint_weights.h5'
    checkpoint_callback = keras.callbacks.ModelCheckpoint(
        filepath=checkpoint_filepath,
        save_weights_only=True,
        monitor='val_accuracy',
        save_best_only=True
    )

    # Reduce LR on plateau
    lr_scheduler = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )

    # Early stopping to prevent overfitting and save compute
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    )

    # Train the model
    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        batch_size=batch_size,
                        epochs=num_epochs,
                        callbacks=[checkpoint_callback, lr_scheduler, early_stopping])

    # Load best weights (redundant if early_stopping restores them, but ensures consistency)
    model.load_weights(checkpoint_filepath)

    # Evaluate on test set
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"Hybrid Model Test Accuracy: {accuracy:.4f}")

    # Optionally save weights and training history
    if save_path != 'none':
        model.save_weights(save_path)
        with open(save_path + "_history.db", 'wb') as file_pi:
            pickle.dump(history.history, file_pi)

    return model, history
```

```python
from sklearn.model_selection import KFold
from tensorflow.keras.models import clone_model

def train_with_kfold(X, y, num_folds=5, batch_size=32, num_epochs=50):
    kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)
    fold_histories = []
    fold_accuracies = []

    for fold, (train_idx, val_idx) in enumerate(kfold.split(X)):
        print(f"\nTraining Fold {fold + 1}/{num_folds}")
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        model = build_hybrid_cnn_vit(input_shape=X.shape[1:], num_classes=y.shape[-1])
        model.compile(optimizer=Adam(learning_rate=1e-3),
                      loss='binary_crossentropy',
                      metrics=['accuracy'])

        callbacks = [
            keras.callbacks.ModelCheckpoint(
                filepath=f'./tmp/fold_{fold+1}_best_weights.h5',
                save_weights_only=True,
                monitor='val_accuracy',
                save_best_only=True
            ),
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1),
            EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
        ]

        history = model.fit(X_train, y_train,
                            validation_data=(X_val, y_val),
                            batch_size=batch_size,
                            epochs=num_epochs,
                            callbacks=callbacks,
                            verbose=2)

        fold_histories.append(history.history)
        loss, acc = model.evaluate(X_val, y_val, verbose=0)
        fold_accuracies.append(acc)
        print(f"Fold {fold + 1} Validation Accuracy: {acc:.4f}")

    print(f"\nMean Validation Accuracy across folds: {np.mean(fold_accuracies):.4f}")
    return fold_histories, fold_accuracies
```
